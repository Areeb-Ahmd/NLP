```python
import re
```

### Extract all the Numbers from Text

```python
text = "Patient's phone is 7321119999. Bill amount is 120$"
pattern = r'\d+' # Any digits repeated n number of times in sequence

match = re.findall(pattern, text)
match
```

### Extract only Phone Number. (Phone no. is always 10 digits)

```python
text = "Patient's phone is 7321119999. Bill amount is 120$"
pattern = r'\d{10}' # Any digits repeated n number of times in sequence

match = re.findall(pattern, text)
match
```

### Extract Phone Number, which is either 10 continuous digits or follows this format (xxx)-xxx-xxx

```python
text = "Patient's phone is (732)-111-9999, spouse phone number 7326664444. Bill amount is 120$"

pattern = r'\(\d{3}\)-\d{3}-\d{4}|\d{10}'

match = re.findall(pattern, text)
match
```

### Extract Phone Number and Bill amount separately

#### Concept of Grouping

```python
text = "Patient's phone is 7321119999. Bill amount is 120$"
pattern = r'(\d{10})\D+(\d+)\$' 

match = re.search(pattern, text)
match
```

```python
match.groups()
```
