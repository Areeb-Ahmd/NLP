<h3>Stemming in NLTK</h3>

```python
from nltk.stem import PorterStemmer #other stemmer -> SnowballStemmer
stemmer = PorterStemmer()
```

```python
words = ["eating", "eats", "eat", "ate", "adjustable", "rafting", "ability", "meeting"]

for word in words:
    print(word, "|", stemmer.stem(word))
```

<h3>Lemmatization in Spacy</h3>

```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")

doc = nlp("eating eats eat ate adjustable rafting ability meeting better")
for token in doc:
    print(token, " | ", token.lemma_) #token.lemma -> prints unique hash of lemma
```

<h3>Customizing lemmatizer</h3>

```python
nlp.pipe_names
```

attribute_ruler -> assigns attribute to a particular token, and we can customize it.

```python
# Customize:
ar = nlp.get_pipe('attribute_ruler') # get attribute_ruler from pipeline

ar.add([[{"TEXT":"Bro"}],[{"TEXT":"Brah"}]],{"LEMMA":"Brother"}) # add custom rule

doc = nlp("Bro, you wanna go? Brah, don't say no! I am exhausted")
for token in doc:
    print(token.text, "|", token.lemma_)
```
